```python
import pandas as pd
```

Import html into pandas dataframe:

```python
alltables = pd.read_html('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M')
```

Identify all tables imported to find required dataframe:

```python
for i in range(0,len(alltables)):
    print ('Table index: ', i, '\n', alltables[i].head(3),'\n')
```

Select required table into dataframe variable (df):

```python
df = alltables[0]
df.shape
```

```python
df.head(20)
```

Ignore cells with a borough that is **Not assigned**:

```python
df.drop(df[df['Borough'] == 'Not assigned'].index, axis = 0, inplace = True)
df.shape
```

```python
df.head(20)
```

If a cell has a borough but a **Not assigned** neighborhood, then the *neighborhood* will be the same as the *borough*.

```python
df['Neighborhood'].replace(to_replace = 'Not assigned', value = df['Borough'], inplace = True)
```

**Dataframe cleanup** (neighborhoods separated by comma, rename column and index reset):

```python
df['Neighborhood']=df['Neighborhood'].apply(lambda x: x.replace(' /', ','))
df.rename(columns={"Postal code": "PostalCode"}, inplace=True)
df.reset_index(inplace=True, drop = True)
df.head()
```
